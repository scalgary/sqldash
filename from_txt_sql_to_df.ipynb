```python
import pandas as pd

# Lire le fichier brut
with open("tables_result.txt", "r") as file:
    lines = file.readlines()

# Supprimer les lignes vides ou décoratives
cleaned_lines = [
    line.strip()
    for line in lines
    if line.strip() and not set(line.strip()) <= {"-", " "}
]

# Supprimer la dernière ligne si elle ne correspond pas aux colonnes
if len(cleaned_lines) >= 2:
    last_line_cols = len(cleaned_lines[-1].split())
    first_line_cols = len(cleaned_lines[0].split())
    if last_line_cols != first_line_cols:
        cleaned_lines = cleaned_lines[:-1]  # Supprimer la dernière ligne

# Extraire les colonnes et données
header = cleaned_lines[1].split()
data = [line.split() for line in cleaned_lines[2:]]


# Créer le DataFrame
df = pd.DataFrame(data, columns=header)

print(df)
```
